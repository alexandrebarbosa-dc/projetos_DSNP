# **Importando as bibliotecas e os dados**

Vamos importar as bibliotecas que serão utilizadas na nossa pesqueisa.

```
# importar as bibliotecas necessárias

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style()
%matplotlib inline

# importar o dataset COVID-19 para um DataaFrame

df = pd.read_csv('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/owid-covid-data.csv')
```

Com essa informação, vamos verificar onde temos o maior número de casos e mortes por COVID até a data do último registro.

```
# Países que têm mais casos e mortes.

df.loc[df.date == '2020-06-15'].sort_values(by="total_cases", ascending=False)
df.loc[df.date == '2020-06-15'].sort_values(by="total_deaths", ascending=False)
```

Assim, podemos verificar que o País com o maior número de casos e de morte em 15 de junho de 2020 é o **Estados Unidos** com 2.094.069 e 115.732 respectivamente. O segundo colocado fica o **Brasil** com 867.624 casos e 43.332 mortes.

Para uma melhor visualização dos dados com relação aos 5 países com maiores números de mortes, vamos plotar o gráfico de barras.

```
# Gráfico do 5 países com maior números de morte

top_5_total_deaths = df.loc[df.date == '2020-06-15' , ['location' , 'total_deaths']].sort_values(by="total_deaths" , ascending=False)[1:6]

fig, ax = plt.subplots(figsize=(10,5))

ax.bar("location" , "total_deaths" , data=top_5_total_deaths)

plt.tight_layout()
```

Aqui podemos visualizar os cinco países com maiores números de mortes. Deixando bastante evidente a grande diferença entre o primeiro colocado que é os Estados Unidos para os demais, tendo esses, uma menor diferença entre eles.

Vamos visualizar como se deu a evolução na quantidade de casos e mortes ao longo do tempo. Essa é uma maneira de observarmos como ocorreu a evolução na quantidade de mortes e de casos, para isso, vamos plotar um gráfico de linha para cada situação.

```
# Evolução do números de mortes ao longo do tempo
fig, ax = plt.subplots(figsize=(10,5))

df.loc[df.location == "World", "total_deaths"].reset_index(drop=True).plot(kind="line" , ax=ax);
ax.set_title("Números de Mortes no Tempo")
ax.set_ylabel("Total de Mortes")
ax.set_xlabel("Dias")

plt.tight_layout()

```

Nesse gráfico, vemos como evolui a quantidade de mortes com o passar dos dias, verificando que a curva de morte tem uma relevante inclinação a partir do 75º dia.

Vamos agora visualizar como se deu a evolução na quantidade de casos ao longo do tempo.

```
# Evolução do números de casos ao longo do tempo

fig, ax = plt.subplots(figsize=(10,5))

df.loc[df.location == "World" , "total_cases"].reset_index(drop=True).plot(kind="line" , ax=ax);
ax.set_title("Números de Casos no Tempo")
ax.set_ylabel("Números de Casos")
ax.set_xlabel("Dias")

plt.tight_layout()
```

De modo semelhante, podemos verificar que o início do aumento no números de casos tem início após o 75º dia.

## **Análise Exploratória para o Brasil**

Vamos agora explorar os dados referentes ao Brasil.
Iniciaremos com a criação de um dataset apenas com os dados referentes ao Brasil.



```
# Dataset do Brasil

df_brasil = df.loc[df.location == "Brazil"].copy()
```

```
# 5 primeiras entradas

df_brasil.head()
```

Vamos verificar qual o intervalo de tempo estão os dados presentes no dataset.
Buscando encontrar qual a data do primeiro e do último registro para os dados referentes ao Brasil.

```
# Menor data
df_brasil.date.min()
```

```
# Maior data

df_brasil.date.max()
```

Podemos concluir que o período de dados existente estão entre os dias **31 de dezembro de 2019** a **15 de junho de 2020**.

Agora, iremos identificar quando ocorreu o primeiro caso e a primeira morte no Brasil. Para isso, vamos localizar os valores diferentes de zero na coluna `new_cases` e solicitar que mostre a primeira entratada nessa situação, assim veremos o primeiro caso de COVID ocorrido no Brasil.

```
# Primeiro casa no Brasil

df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
```

Logo, temos que a primeira o primeiro caso de COVID registrado no Brasil ocorreu no dia **26 de fevereiro de 2020.**

Já que temos a data do primeiro caso registrado, vamos descobrir a data da primeira morte registrada e vermos quanto tempo ocorreu entre o primeiro caso e a primeira morte.

Vamos proceder da mesma maneira, encontrando inicialmente os registros deferentes de zero na coluna`new_deaths` e solicitando que seja mostrado o primeiro valor.

```
# Primeira morte de COVID resgitrada no Brasil
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)
```

Chegamos então a data de registro da primeira morte registrada por COVID no Brasil sendo no dia **18 de março de 2020.**

Podemos concluir que entre o registro do primeiro caso **(26 de fevereiro de 2020)** e a primeira morte **(18 de março de 2020)** se passaram **21 dias.**

Vamos plotar os gráficos de linhas para observarmos o comportamento da evolução de casos e mortes no Brasil.

```
# Gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais no Brasil.

x = list(range(168))
y = df_brasil['total_cases']
y2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(11, 9))

ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, y)

ax2.set_title('Casos no Brasil (Escala Logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x ,y)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, y2)

ax4.set_title('Mortes no Brasil (Escala Logarítmica)')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, y2)
plt.tight_layout()
```

De modo semalhante ao observado no mundo, a quantidade de casos e mortes no Brasil se intensifica a partir do 75º dia, como podemos observar nos gráficos.
