Através dessa plataforma, as pessoas podem alugar sua própria casa, apartamento ou quarto extra para viajantes em busca de autônomos e acolhedoras.

A história do Airbnb começou quando os fundadores, Brian Chesky e Joe Gebbia, decidiram abrir as portas de seu próprio apartamento em San Francisco para ajudar a pagar o aluguel. Eles criaram um site simples para oferecer colchões infláveis ​​aos viajantes que estavam com dificuldades em encontrar acomodação durante uma conferência na cidade. A ideia foi um sucesso e logo perceberam que havia uma demanda crescente por esse tipo de hospedagem alternativa.

Em 2009, eles foram acompanhados por Nathan Blecharczyk, que se tornou o terceiro cofundador. Juntos, eles expandiram o conceito do Airbnb e lançaram a plataforma oficialmente. O Airbnb rapidamente ganhou popularidade e se seguiu para outras cidades dos Estados Unidos e, posteriormente, para todo o mundo.

A plataforma do Airbnb permite que as pessoas anunciem suas propriedades disponíveis para aluguel de curto prazo. Os hóspedes podem criar perfis, adicionar fotos e vídeos ao vivo de suas acomodações, além de definir preços e regras para os hóspedes. Por sua vez, os viajantes podem pesquisar e reservar com base em suas comensais, dados de viagem e orçamento.

O Airbnb oferece uma ampla variedade de opções de hospedagem, desde quartos privativos em residências compartilhadas até casas inteiras, apartamentos e até mesmo acomodações únicas, como cabanas, barcos e castelos. Os viajantes também têm a oportunidade de se conectar com os hóspedes, obter recomendações locais e ter uma experiência mais imersiva na cidade que estão visitando.

Além disso, o Airbnb oferece um sistema de avaliação e comentários, o que permite que os usuários compartilhem suas experiências e ajudem a garantir a confiança e a segurança nas transações. A empresa também implementou medidas de segurança e verificação de identidade para proteger os proprietários e os hóspedes.

Com o passar dos anos, o Airbnb expandiu seu alcance globalmente, permanecendo presente em mais de 220 países e regiões ao redor do mundo. A plataforma tem desempenhado um papel significativo no setor de viagens, fornecendo uma alternativa flexível e acessível às opções de hospedagem tradicionais.

Em resumo, o Airbnb revolucionou a forma como as pessoas viajam e se hospedam, oferecendo uma maneira única de conectar hóspedes e viajantes, promovendo experiências registradas e personalizadas. Sua história de sucesso é um testemunho do poder da economia compartilhada e da crescente demanda por hóspedes únicos e memoráveis ​​em todo o mundo.

###**Plotando o Bloxplot**

###**Histogramas sem outliers**
Já que identificamos outliers nas variáveis `price` e `minimum_nights`, vamos agora limpar o DataFrame delas e plotar novamente o histograma.
Optamos por eliminar os valores acima de 100 noites para a entrada que se refere a quantidade mínima de noites e os valores acima de 500 dólares canadenses para o preço por noite.

```
# remover os *outliers* em um novo DataFrame
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 500].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 100].index, axis=0, inplace=True)

# plotar o histograma para as variáveis numéricas
df_clean.hist(bins=15, figsize=(15,10));
```

###**Vamos verificar o resumo estatístico após o tratamento dos dados**

```
# ver o resumo estatístico das variáveis numéricas após a limpeza
df_clean[['price', 'minimum_nights', 'number_of_reviews', 'calculated_host_listings_count'
 , 'availability_365']].describe()
```

###**Verificando a média dos preços de aluguel após a limpeza dos 'outliers'**

```
# ver a média da coluna `price``
df_clean.price.mean()
```

Verificamos que a média de preço por noite diminuiu após a limpeza dos 'outliers', onde antes estava em 165.82 doláres canadenses e agora está em 123.88 dólares canadenses.

### **Vamos verificar qual a correlação existente entre as variáveis**

```
# criar uma matriz de correlação
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews',
    'calculated_host_listings_count', 'availability_365']].corr()
# mostrar a matriz de correlação
display(corr)
```

```
# plotar um heatmap a partir das correlações
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

Podemos notar uma maior correlação positiva entre o `price`e o `number_of_reviews`, de modo que podemos esperar que com os preços maiores, a qualidade do imóvel aumenta e consequentemente a quantidade de avaliações.

Com relação a correlação negativa, temos uma maior em relação ao `minimum_nights`e o `price`, isso nos mostra que não há uma relação direta entre um maior valor do imóvel com relação a quantidade de noite mínima para locação.

Contudo, não notamos uma correlação tão significativa entre as variáveis.

A quantidade de noites mínimas para alugar é de 19 noites em médias, porém, 75% dos imóveis tem a quantidade de noites mínima de 31 noites, contudo, também temos muitos imóveis com quantidade mínima de 2 dias, com percentual de 25% dos imóveis analisados.

Foi demonstrado que temos aproximadamente 83% dos imóveis disponíveis para aluguel em Montreal são do tipo apartamento ou casas inteiras, o que justifica a média de 19 noites mínima para alugar, 16% dos imóveis são quartos privados, que podem ser alugados por uma menor quantidade de noites.

Com relação ao preço por localidade dos imóveis, após a limpeza dos dados e definindo como maior valor de aluguel em 500 dólares canadenses, temos que a região com maior média de preço fica em na ilha de L'Île-Bizard, com média de 236 dólares. Levando em consideração o valor máximo de preço encontrado que foi de 108546 dólares, temos a média de preço de 525 dólares na localidade de Kirkland.

Uma coriosidade observada é a localização dos dois lugares que encontramos com maior valor médio dos preços, pois ambos não estão localizados na região central da cidade, mais sim afastada do grande centro.
